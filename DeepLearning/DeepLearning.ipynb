 # El Perceptrón: La Unidad Básica del Deep Learning
 
 El perceptrón es el bloque fundamental de las redes neuronales artificiales. Inspirado en las neuronas biológicas, el perceptrón toma múltiples entradas, les asigna pesos, y produce una salida basada en una función de activación.
 
 ### Componentes principales:
 
 1. **Entradas (x)**: Los datos que recibe el perceptrón
 2. **Pesos (w)**: Valores que determinan la importancia de cada entrada
 3. **Sesgo (b)**: Un término adicional que ayuda al modelo a ajustarse mejor
 4. **Función de suma ponderada**: Multiplica cada entrada por su peso y suma los resultados
 5. **Función de activación**: Transforma la suma ponderada en la salida final
 
 ### Funcionamiento:
 
 1. Recibe las entradas (x₁, x₂, ..., xₙ)
 2. Multiplica cada entrada por su peso correspondiente (w₁, w₂, ..., wₙ)
 3. Suma todos los productos y añade el sesgo: Σ(wᵢxᵢ) + b
 4. Aplica la función de activación para obtener la salida
 
 Las funciones de activación más comunes se muestran en las gráficas siguientes, cada una con sus propias características y casos de uso.


```python
import numpy as np
import matplotlib.pyplot as plt

# Definir un perceptrón simple para clasificación binaria
class Perceptron:
    def __init__(self):
        self.w = np.random.randn(2)  # Pesos aleatorios para 2 características
        self.b = np.random.randn(1)  # Sesgo aleatorio
    
    def forward(self, x):
        # Suma ponderada
        z = np.dot(x, self.w) + self.b
        # Función de activación (paso)
        return 1 if z > 0 else 0

# Generar datos de ejemplo
np.random.seed(0)
n_points = 100

# Crear dos grupos de puntos
grupo1_x = np.random.normal(2, 1, (n_points//2, 2))  # Centrado en (2,2)
grupo2_x = np.random.normal(-2, 1, (n_points//2, 2))  # Centrado en (-2,-2)
X = np.vstack([grupo1_x, grupo2_x])
y = np.array([1]*50 + [0]*50)

# Crear y configurar el gráfico
plt.figure(figsize=(12, 8))

# Graficar los puntos
plt.scatter(grupo1_x[:, 0], grupo1_x[:, 1], c='blue', label='Clase 1', alpha=0.6)
plt.scatter(grupo2_x[:, 0], grupo2_x[:, 1], c='red', label='Clase 0', alpha=0.6)

# Crear un perceptrón y visualizar la frontera de decisión
perceptron = Perceptron()
x_decision = np.linspace(-4, 4, 100)
y_decision = -(perceptron.w[0]*x_decision + perceptron.b)/perceptron.w[1]

plt.plot(x_decision, y_decision, 'g--', label='Frontera de decisión')

# Añadir elementos visuales explicativos
plt.annotate('w₁x₁ + w₂x₂ + b = 0', 
            xy=(0, 0), xytext=(1, 2),
            arrowprops=dict(facecolor='black', shrink=0.05))

plt.title('Perceptrón: Clasificación Binaria', fontsize=14)
plt.xlabel('Característica 1 (x₁)', fontsize=12)
plt.ylabel('Característica 2 (x₂)', fontsize=12)
plt.legend()
plt.grid(True, alpha=0.3)
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)

# Añadir texto explicativo
plt.text(-4, 4, 'El perceptrón define una\nfrontera de decisión lineal\nque separa las dos clases', 
         bbox=dict(facecolor='white', alpha=0.7))

plt.show()
```

# Funciones de Activación y Estructura del Perceptrón

Las funciones de activación son componentes fundamentales en las redes neuronales y el perceptrón. Determinan la salida de una neurona en base a sus entradas ponderadas.

## Funciones de Activación Comunes:

1. **Función Lineal**: Es la más simple, donde la salida es proporcional a la entrada (y = x)
2. **ReLU (Rectified Linear Unit)**: Devuelve 0 para entradas negativas y la entrada misma para valores positivos
3. **Sigmoid**: Mapea cualquier entrada a un valor entre 0 y 1, útil para clasificación binaria
4. **Tanh**: Similar a sigmoid pero mapea valores entre -1 y 1

## Estructura del Perceptrón:

El perceptrón es la unidad básica de una red neuronal y consiste en:

1. **Entradas (x₁, x₂, ..., xₙ)**: Los datos de entrada
2. **Pesos (w₁, w₂, ..., wₙ)**: Parámetros ajustables que determinan la importancia de cada entrada
3. **Sesgo (b)**: Un término adicional que permite ajustar el umbral de activación
4. **Función de Activación**: Determina la salida final del perceptrón

La salida se calcula como: f(w₁x₁ + w₂x₂ + ... + wₙxₙ + b), donde f es la función de activación.


```python
import numpy as np
import matplotlib.pyplot as plt

# Crear datos para el eje x
x = np.linspace(-10, 10, 200)

# Definir las funciones de activación
linear = x
relu = np.maximum(0, x)
sigmoid = 1 / (1 + np.exp(-x))
tanh = np.tanh(x)

# Crear subplots
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(12, 8))

# Graficar función lineal
ax1.plot(x, linear, color='blue')
ax1.set_title('Función Lineal')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.grid(True, linestyle='--', alpha=0.7)
ax1.axhline(y=0, color='k', linestyle='-', alpha=0.3)
ax1.axvline(x=0, color='k', linestyle='-', alpha=0.3)

# Graficar función ReLU
ax2.plot(x, relu, color='red')
ax2.set_title('Función ReLU')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.grid(True, linestyle='--', alpha=0.7)
ax2.axhline(y=0, color='k', linestyle='-', alpha=0.3)
ax2.axvline(x=0, color='k', linestyle='-', alpha=0.3)

# Graficar función Sigmoid
ax3.plot(x, sigmoid, color='green')
ax3.set_title('Función Sigmoid')
ax3.set_xlabel('x')
ax3.set_ylabel('y')
ax3.grid(True, linestyle='--', alpha=0.7)
ax3.axhline(y=0, color='k', linestyle='-', alpha=0.3)
ax3.axvline(x=0, color='k', linestyle='-', alpha=0.3)

# Graficar función Tanh
ax4.plot(x, tanh, color='purple')
ax4.set_title('Función Tanh')
ax4.set_xlabel('x')
ax4.set_ylabel('y')
ax4.grid(True, linestyle='--', alpha=0.7)
ax4.axhline(y=0, color='k', linestyle='-', alpha=0.3)
ax4.axvline(x=0, color='k', linestyle='-', alpha=0.3)

# Ajustar el espaciado entre subplots
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

# Api secuencial

La API Secuencial de TensorFlow es una de las formas más intuitivas y sencillas para construir modelos de deep learning, especialmente cuando se trata de redes neuronales en las que las capas se organizan de forma lineal, una detrás de otra. A continuación, se presenta una descripción detallada de esta API, sus características y cómo se utiliza en el desarrollo de modelos de aprendizaje profundo.
¿Qué es la API Secuencial?

La API Secuencial es una abstracción de alto nivel dentro de TensorFlow (especialmente en Keras, la biblioteca de alto nivel integrada en TensorFlow) que permite crear modelos de redes neuronales de manera rápida y sencilla. Con ella, se pueden apilar capas de forma secuencial, lo que es ideal para modelos en los que cada capa tiene exactamente una entrada y una salida.
Características Principales

- Simplicidad y Facilidad de Uso:
La API Secuencial permite definir modelos con pocas líneas de código. Cada capa se añade en el orden en que se desea que los datos fluyan a través de la red.

- Estructura Lineal:
Está diseñada para modelos donde la estructura es lineal, es decir, donde las capas se encadenan de manera directa sin bifurcaciones o combinaciones complejas. Esto la hace perfecta para redes neuronales densas (fully connected) y modelos simples de redes convolucionales o recurrentes.

- Integración con Keras:
Como parte de Keras, la API Secuencial se beneficia de todas las herramientas y funciones que Keras ofrece, incluyendo la compilación del modelo, la definición de funciones de pérdida, optimizadores y métricas de evaluación.

- Extensibilidad:
Aunque es una API sencilla, se puede extender mediante la combinación con otras funcionalidades de Keras. Para arquitecturas más complejas o modelos que requieran conexiones múltiples, TensorFlow ofrece la API funcional, pero para muchos casos prácticos la API Secuencial es suficiente.

 
 En el aprendizaje profundo, los optimizadores y las funciones de pérdida son componentes cruciales para el entrenamiento de modelos de redes neuronales.
 
 ### Optimizadores
 
 Los optimizadores son algoritmos o métodos que cambian los atributos de la red neuronal, como los pesos y las tasas de aprendizaje, con el fin de reducir las pérdidas. Algunos de los optimizadores más comunes son:
 
 - **SGD (Stochastic Gradient Descent)**: Es uno de los optimizadores más simples y comunes. Actualiza los pesos de la red en función del gradiente de la función de pérdida con respecto a los pesos.
 - **Adam (Adaptive Moment Estimation)**: Combina las ventajas de dos otros métodos: AdaGrad y RMSProp. Es eficiente en términos de memoria y generalmente funciona bien en una amplia variedad de problemas.
 - **RMSProp (Root Mean Square Propagation)**: Ajusta la tasa de aprendizaje para cada uno de los parámetros. Es especialmente útil para problemas de redes neuronales recurrentes.
 
 ### Funciones de Pérdida

 Las funciones de pérdida cuantifican la discrepancia entre las predicciones del modelo y los valores reales, proporcionando una métrica que el optimizador minimiza durante el entrenamiento. La elección de la función de pérdida depende del tipo de problema que se está abordando. Algunas de las más comunes son:
Para problemas de regresión:

- Error Cuadrático Medio (MSE - Mean Squared Error): Calcula la media de los cuadrados de los errores. Penaliza más los errores grandes, lo que puede hacer que sea sensible a valores atípicos.
- Error Absoluto Medio (MAE - Mean Absolute Error): Calcula la media de los valores absolutos de los errores. Es más robusto ante valores atípicos que el MSE.
- Huber Loss: Combina MSE y MAE, usando MSE para errores pequeños y MAE para errores grandes, lo que la hace menos sensible a valores atípicos.

Para problemas de clasificación:

-Entropía Cruzada (Cross-Entropy Loss o Log Loss): Se utiliza en clasificación y mide la diferencia entre la distribución real y la predicha. Es común en modelos de clasificación binaria y multiclase.
    - Binary Cross-Entropy: Se usa para clasificación binaria.
    - Categorical Cross-Entropy: Se usa para clasificación multiclase con etiquetas one-hot.
    - Sparse Categorical Cross-Entropy: Similar a la anterior, pero se usa cuando las etiquetas son enteros en lugar de one-hot.

Para problemas de segmentación y detección:

- IoU Loss (Intersection over Union Loss): Se usa en visión por computadora para medir la superposición entre una predicción y la verdad de referencia.
- Dice Loss: Similar a IoU, se usa en segmentación de imágenes para medir la coincidencia entre la máscara predicha y la real.

La elección del optimizador y la función de pérdida adecuados es fundamental para el éxito del entrenamiento de redes neuronales profundas, ya que afectan la velocidad de convergencia y la precisión del modelo.

### Entrenamiento de Redes Neuronales y Backpropagation  

El entrenamiento de redes neuronales es el proceso mediante el cual el modelo aprende a ajustar sus pesos y sesgos para minimizar el error en las predicciones. Este proceso se basa en la retropropagación del error (*backpropagation*) y la optimización de los parámetros mediante algoritmos como *Stochastic Gradient Descent (SGD)* o *Adam*.  

### Backpropagation  

Backpropagation, o retropropagación del error, es un algoritmo fundamental para el aprendizaje profundo. Su objetivo es calcular de manera eficiente los gradientes de la función de pérdida con respecto a los pesos de la red neuronal, permitiendo su ajuste mediante optimización. El proceso de backpropagation consta de los siguientes pasos:  

1. **Forward Pass**: Se propaga la entrada a través de la red, pasando por cada capa, hasta obtener una salida predicha.  
2. **Cálculo de la Pérdida**: Se compara la salida del modelo con la salida real utilizando una función de pérdida.  
3. **Backward Pass (Retropropagación del Error)**: Se calculan los gradientes de la función de pérdida con respecto a los parámetros de la red usando la regla de la cadena. Estos gradientes indican cómo cambiar los pesos para reducir el error.  
4. **Actualización de Pesos**: Con los gradientes calculados, un optimizador (como SGD o Adam) ajusta los pesos de la red en la dirección que minimiza la pérdida.  

Este proceso se repite durante múltiples iteraciones, denominadas *épocas*, hasta que la función de pérdida alcanza un valor óptimo o deja de mejorar significativamente. La eficacia del entrenamiento depende de factores como la tasa de aprendizaje, el tamaño del lote (*batch size*) y la arquitectura de la red.  

Backpropagation ha sido clave en el desarrollo de redes neuronales profundas, permitiendo el entrenamiento eficiente de modelos complejos en tareas como reconocimiento de imágenes, procesamiento de lenguaje natural y juegos de inteligencia artificial.

# Redes neuronales convolucionales

```python
# Ejemplo de una red neuronal convolucional (CNN) usando Keras

from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

# Definir el modelo
cnn_model = Sequential()

# Añadir una capa convolucional con 32 filtros, tamaño de kernel 3x3, función de activación ReLU y tamaño de entrada 28x28x1
cnn_model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)))

# Añadir una capa de max-pooling con tamaño de pool 2x2
cnn_model.add(MaxPooling2D((2, 2)))

# Añadir una segunda capa convolucional con 64 filtros y tamaño de kernel 3x3
cnn_model.add(Conv2D(64, (3, 3), activation='relu'))

# Añadir una segunda capa de max-pooling con tamaño de pool 2x2
cnn_model.add(MaxPooling2D((2, 2)))

# Aplanar las salidas de las capas anteriores para conectarlas a una capa densa
cnn_model.add(Flatten())

# Añadir una capa densa con 64 unidades y función de activación ReLU
cnn_model.add(Dense(64, activation='relu'))

# Añadir una capa de salida con 10 unidades (para 10 clases) y función de activación softmax
cnn_model.add(Dense(10, activation='softmax'))

# Compilar el modelo
cnn_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Resumen del modelo
cnn_model.summary()
```

```python
# Train the model
history= cnn_model.fit(x_train, y_train, epochs=5, batch_size=128, validation_split=0.2)
```

```python
# Batch size
BATCH_SIZE = 64

# Buffer size to shuffle the dataset
# (TF data is designed to work with possibly infinite sequences,
# so it doesn't attempt to shuffle the entire sequence in memory. Instead,
# it maintains a buffer in which it shuffles elements).
BUFFER_SIZE = 10000

dataset = (
    dataset
    .shuffle(BUFFER_SIZE)
    .batch(BATCH_SIZE, drop_remainder=True)
    .prefetch(tf.data.experimental.AUTOTUNE))

dataset
```

### Build the model

```python
# Length of the vocabulary in StringLookup Layer
vocab_size = len(ids_from_chars.get_vocabulary())

# The embedding dimension
embedding_dim = 256

# Number of RNN units
rnn_units = 1024
```

```python
class MyModel(tf.keras.Model):
    def __init__(self, vocab_size, embedding_dim, rnn_units):
        super().__init__()
        self.embedding = tf.keras.layers.Embedding(vocab_size, embedding_dim)
        self.gru = tf.keras.layers.GRU(rnn_units,
                                    return_sequences=True,
                                    return_state=True)
        self.dense = tf.keras.layers.Dense(vocab_size)

    def call(self, inputs, states=None, return_state=False, training=False):
        x = inputs
        x = self.embedding(x, training=training)
        if states is None:
            # Corregir la obtención del estado inicial
            states = tf.zeros([tf.shape(x)[0], self.gru.units])
        x, states = self.gru(x, initial_state=states, training=training)
        x = self.dense(x, training=training)

        if return_state:
            return x, states
        else:
            return x
```

```python
model = MyModel(
    vocab_size=vocab_size,
    embedding_dim=embedding_dim,
    rnn_units=rnn_units)
```

```python
for input_example_batch, target_example_batch in dataset.take(1):
    example_batch_predictions = model(input_example_batch)
    print(example_batch_predictions.shape, "# (batch_size, sequence_length, vocab_size)")
```

```python
class OneStep(tf.keras.Model):
  def __init__(self, model, chars_from_ids, ids_from_chars, temperature=1.0):
    super().__init__()
    self.temperature = temperature
    self.model = model
    self.chars_from_ids = chars_from_ids
    self.ids_from_chars = ids_from_chars

    # Create a mask to prevent "[UNK]" from being generated.
    skip_ids = self.ids_from_chars(['[UNK]'])[:, None]
    sparse_mask = tf.SparseTensor(
        # Put a -inf at each bad index.
        values=[-float('inf')]*len(skip_ids),
        indices=skip_ids,
        # Match the shape to the vocabulary
        dense_shape=[len(ids_from_chars.get_vocabulary())])
    self.prediction_mask = tf.sparse.to_dense(sparse_mask)

  @tf.function
  def generate_one_step(self, inputs, states=None):
    # Convert strings to token IDs.
    input_chars = tf.strings.unicode_split(inputs, 'UTF-8')
    input_ids = self.ids_from_chars(input_chars).to_tensor()

    # Run the model.
    # predicted_logits.shape is [batch, char, next_char_logits]
    predicted_logits, states = self.model(inputs=input_ids, states=states,
                                          return_state=True)
    # Only use the last prediction.
    predicted_logits = predicted_logits[:, -1, :]
    predicted_logits = predicted_logits/self.temperature
    # Apply the prediction mask: prevent "[UNK]" from being generated.
    predicted_logits = predicted_logits + self.prediction_mask

    # Sample the output logits to generate token IDs.
    predicted_ids = tf.random.categorical(predicted_logits, num_samples=1)
    predicted_ids = tf.squeeze(predicted_ids, axis=-1)

    # Convert from token ids to characters
    predicted_chars = self.chars_from_ids(predicted_ids)

    # Return the characters and model state.
    return predicted_chars, states
```

```python
one_step_model = OneStep(model, chars_from_ids, ids_from_chars)
```

```python
start = time.time()
states = None
next_char = tf.constant(['Juliette : Romeo I love you'])
result = [next_char]

for n in range(1000):
  next_char, states = one_step_model.generate_one_step(next_char, states=states)
  result.append(next_char)

result = tf.strings.join(result)
end = time.time()
print(result[0].numpy().decode('utf-8'), '\n\n' + '_'*80)
print('\nRun time:', end - start)
```
