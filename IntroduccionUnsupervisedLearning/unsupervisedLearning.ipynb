 ## Introducción 

El **aprendizaje no supervisado** es una rama fundamental del aprendizaje automático que se centra en explorar y analizar datos sin necesidad de etiquetas o información previa sobre los resultados esperados. A diferencia del aprendizaje supervisado, donde los algoritmos se entrenan con datos etiquetados para realizar predicciones o clasificaciones específicas, el aprendizaje no supervisado permite que el modelo descubra patrones ocultos, estructuras subyacentes y relaciones inherentes dentro de los datos.  

Este enfoque es particularmente valioso cuando la etiquetación de datos es costosa, complicada o incluso imposible, lo que lo hace aplicable a una amplia variedad de dominios y contextos.  

### Diferencias clave con el aprendizaje supervisado  

El aprendizaje no supervisado tiene características distintivas que lo separan del aprendizaje supervisado:  
- **Ausencia de etiquetas**: No existen "respuestas correctas" o etiquetas predefinidas que guíen al modelo.  
- **Objetivo de descubrimiento**: En lugar de realizar predicciones específicas, el objetivo principal es descubrir la estructura y organización subyacentes de los datos.  
- **Evaluación subjetiva**: La calidad de los resultados depende en gran medida de la interpretación humana y del contexto del problema, en lugar de métricas de precisión claramente definidas.  
- **Menor preparación de datos**: Requiere menos esfuerzo para etiquetar datos, pero demanda un análisis más profundo y cuidadoso de los resultados obtenidos para extraer conclusiones significativas.  

### Aplicaciones comunes del aprendizaje no supervisado  

El aprendizaje no supervisado es una herramienta poderosa que se aplica en diversos sectores e industrias para resolver problemas complejos y analizar grandes volúmenes de datos. Algunas de sus aplicaciones más comunes incluyen:  

1. **Segmentación de clientes**:  
   Agrupación de clientes en segmentos homogéneos basados en comportamientos, preferencias o características similares, lo que facilita estrategias de marketing personalizadas y la optimización de productos.  

2. **Detección de anomalías**:  
   Identificación de datos atípicos que podrían representar fraudes, errores de medición o comportamientos inusuales en áreas como seguridad informática, sistemas financieros y mantenimiento predictivo.  

3. **Sistemas de recomendación**:  
   Uso de patrones de datos para sugerir productos, películas, música o contenido personalizado basado en preferencias pasadas y similitudes con otros usuarios.  

4. **Agrupamiento de documentos**:  
   Organización de grandes volúmenes de texto, como artículos o publicaciones, en grupos temáticos para facilitar la búsqueda y el análisis.  

5. **Análisis de redes sociales**:  
   Identificación de comunidades, influencers y patrones de interacción entre usuarios dentro de redes sociales, permitiendo una comprensión más profunda de dinámicas sociales y relaciones.  

6. **Compresión de datos**:  
   Reducción de la dimensionalidad de conjuntos de datos complejos para simplificar su almacenamiento, procesamiento y análisis, preservando la información más relevante.  

7. **Detección de patrones en datos biomédicos**:  
   Descubrimiento de relaciones complejas en datos médicos y biológicos, como genes asociados a enfermedades, patrones en imágenes médicas y resultados de experimentos clínicos.  

### Relevancia del aprendizaje no supervisado  

En un mundo impulsado por el crecimiento exponencial de los datos, el aprendizaje no supervisado se ha convertido en una herramienta esencial para transformar datos no estructurados en conocimientos valiosos. Desde mejorar la experiencia del cliente hasta acelerar descubrimientos científicos, su capacidad para revelar patrones ocultos es crucial en la toma de decisiones basada en datos.  

Aunque presenta desafíos, como la interpretación y evaluación de resultados, su flexibilidad y versatilidad lo convierten en un pilar fundamental de la inteligencia artificial y la ciencia de datos moderna.



### 2. Algoritmos de Clustering  



## Algoritmo de K-means:

El **algoritmo K-means** es uno de los métodos más utilizados y sencillos para realizar agrupamiento (clustering) en conjuntos de datos no etiquetados. Su objetivo principal es particionar los datos en un número predefinido de clústeres (K) de manera que los puntos dentro de un mismo clúster sean lo más similares posible entre sí, y lo más diferentes posible de los puntos en otros clústeres.  

K-means es ampliamente utilizado debido a su simplicidad y eficiencia computacional, siendo aplicable en dominios como la segmentación de clientes, la compresión de imágenes y el análisis de patrones.

---

### Cómo funciona el algoritmo K-means  

El algoritmo sigue un proceso iterativo para asignar puntos de datos a clústeres y ajustar los centroides de cada clúster. Su funcionamiento básico se divide en los siguientes pasos:  

#### **1. Inicialización**  
Se seleccionan aleatoriamente K centroides iniciales, donde K es el número de clústeres que se desea crear. Los centroides son puntos en el espacio que representan el centro de los clústeres.  

#### **2. Asignación de puntos**  
Cada punto de datos se asigna al clúster cuyo centroide esté más cercano, utilizando una función de distancia, como la distancia euclidiana. Este paso crea K grupos iniciales.  

#### **3. Actualización de centroides**  
Para cada clúster, se recalcula el centroide como el promedio de todos los puntos asignados a ese clúster.  

#### **4. Repetición**  
Los pasos de asignación de puntos y actualización de centroides se repiten hasta que:  
- Las asignaciones de los puntos a los clústeres ya no cambien.  
- La variación en las posiciones de los centroides entre iteraciones sea mínima.  
- Se alcance un número máximo de iteraciones definido previamente.  

---

### Planteamiento formal

El algoritmo intenta minimizar una función objetivo llamada **inercia intra-clúster**, que se define como la suma de las distancias cuadradas entre cada punto y el centroide de su clúster:  

$$
J = \sum_{i=1}^{K} \sum_{x \in C_i} \| x - \mu_i \|^2
$$  

Donde:  
- $ K $: Número de clústeres.  
- $ C_i $: Conjunto de puntos en el clúster $ i $.  
- $ \mu_i $: Centroide del clúster $ i $.  
- $ x $: Punto de datos en el clúster $ C_i $.  

El objetivo del algoritmo es minimizar $ J $, lo que equivale a encontrar una partición óptima de los datos.  

---

### Ventajas del algoritmo K-means  

1. **Simplicidad y rapidez**:  
   Es fácil de entender e implementar. Su complejidad computacional es aproximadamente $ O(n \cdot K \cdot t) $, donde $ n $ es el número de puntos, $ K $ es el número de clústeres y $ t $ el número de iteraciones.  

2. **Escalabilidad**:  
   Es eficiente con conjuntos de datos grandes y escalable a problemas con dimensiones moderadas.  

3. **Versatilidad**:  
   Puede aplicarse en una amplia variedad de problemas, desde la segmentación de clientes hasta la compresión de datos.  

---

### Limitaciones del algoritmo K-means  

1. **Selección de K**:  
   Requiere definir el número de clústeres a priori, lo cual puede ser un desafío si no se tiene conocimiento previo sobre los datos. Una técnica común para determinar $ K $ es el método del **codo** (Elbow Method).  

2. **Sensibilidad a los valores atípicos**:  
   Puntos de datos atípicos o ruido pueden afectar significativamente la posición de los centroides.  

3. **Formas de los clústeres**:  
   Supone que los clústeres tienen formas esféricas y tamaños similares, lo que puede ser una limitación en conjuntos de datos con clústeres de formas irregulares o tamaños desiguales.  

4. **Dependencia de la inicialización**:  
   Diferentes inicializaciones de los centroides pueden conducir a diferentes resultados, por lo que el algoritmo puede quedarse atascado en mínimos locales. Una mejora común es usar el método **K-means++** para inicializar mejor los centroides.  

```python
# Importamos las bibliotecas necesarias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Generamos datos de ejemplo
np.random.seed(42)
n_samples = 300

# Creamos dos características: monto_gastado y frecuencia_compra
monto_gastado = np.random.normal(1000, 300, n_samples)
frecuencia_compra = np.random.normal(10, 3, n_samples)

# Creamos un DataFrame con los datos
data = pd.DataFrame({
    'monto_gastado': monto_gastado,
    'frecuencia_compra': frecuencia_compra
})

```

```python
data
```

```python
# Estandarizamos los datos
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

# Aplicamos K-means
n_clusters = 4
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
clusters = kmeans.fit_predict(data_scaled)

# Añadimos las etiquetas de cluster al DataFrame
data['Cluster'] = clusters

# Visualizamos los resultados
plt.figure(figsize=(10, 6))
scatter = plt.scatter(data['monto_gastado'], data['frecuencia_compra'], 
                     c=data['Cluster'], cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:,0]*scaler.scale_[0] + scaler.mean_[0],
            kmeans.cluster_centers_[:,1]*scaler.scale_[1] + scaler.mean_[1],
            marker='x', s=200, linewidths=3, color='r', label='Centroides')
plt.xlabel('Monto Gastado ($)')
plt.ylabel('Frecuencia de Compra')
plt.title('Segmentación de Clientes usando K-means')
plt.legend()
plt.colorbar(scatter, label='Cluster')
plt.show()

# Analizamos los centroides
centroids_original = scaler.inverse_transform(kmeans.cluster_centers_)
for i in range(n_clusters):
    print(f"\nCluster {i}:")
    print(f"Monto promedio: ${centroids_original[i,0]:.2f}")
    print(f"Frecuencia promedio: {centroids_original[i,1]:.2f}")
```



## 2. Agrupamiento Jerárquico: Una visión estructurada del clustering

El **agrupamiento jerárquico** es un enfoque de clustering que organiza los datos en una estructura jerárquica similar a un árbol, llamada **dendrograma**. Este método permite visualizar las relaciones entre los datos de una manera intuitiva, facilitando la comprensión de los patrones subyacentes. A diferencia de K-means, no requiere especificar el número de clústeres $ K $ de antemano, ya que construye una estructura que permite seleccionar $ K $ después de analizar el dendrograma.

### Cómo funciona el Agrupamiento Jerárquico Aglomerativo

El proceso se basa en medir la similitud o disimilitud entre los puntos de datos y agruparlos iterativamente. Los pasos básicos son:

#### **1. Inicialización**  
Cada punto de datos se considera un clúster independiente. En un conjunto con $ n $ puntos, se comienza con $ n $ clústeres.  

#### **2. Cálculo de distancias**  
Se calcula la distancia entre todos los clústeres. Las medidas de distancia más comunes incluyen:  
- **Distancia simple (Single Linkage):** Distancia mínima entre puntos de diferentes clústeres.  
- **Distancia completa (Complete Linkage):** Distancia máxima entre puntos de diferentes clústeres.  
- **Promedio (Average Linkage):** Promedio de las distancias entre puntos de diferentes clústeres.  
- **Distancia del centroide:** Distancia entre los centroides de los clústeres.  

#### **3. Combinación de clústeres**  
Los dos clústeres con menor distancia (más similares) se combinan en un nuevo clúster.  

#### **4. Actualización de distancias**  
Se recalculan las distancias entre los nuevos clústeres formados y el resto.  

#### **5. Repetición**  
Se repiten los pasos 3 y 4 hasta que todos los puntos formen un único clúster o se alcance el número deseado de clústeres.  



### Representación mediante un Dendrograma

Un dendrograma es una representación visual del agrupamiento jerárquico.  
- Cada hoja del dendrograma representa un punto de datos.  
- Las bifurcaciones muestran cómo los clústeres se combinan.  
- La altura de las bifurcaciones indica la distancia o disimilitud entre los clústeres.  
El usuario puede seleccionar el número de clústeres cortando el dendrograma en un nivel deseado.


## Formalización del Clustering Jerárquico

El clustering jerárquico se basa en el uso de **matrices de distancias** y **métodos de enlace** para calcular las similitudes entre los datos y fusionarlos en clústeres de forma iterativa. A continuación, se explican los aspectos matemáticos clave:


### 1. **Definición de Distancia**

El primer paso en el clustering jerárquico es calcular la distancia o disimilitud entre los puntos de datos. Para un conjunto de datos con $ n $ puntos en $ \mathbb{R}^m $, las distancias se representan mediante una **matriz de distancia** $ D $, de tamaño $ n \times n $, donde cada entrada $ d_{ij} $ es la distancia entre los puntos $ i $ y $ j $.

#### **Medidas comunes de distancia**:
1. **Distancia Euclidiana**:
   $$
   d_{ij} = \sqrt{\sum_{k=1}^{m} (x_{ik} - x_{jk})^2}
   $$
   Es la métrica más común y mide la distancia "recta" entre dos puntos.

2. **Distancia Manhattan (L1)**:
   $$
   d_{ij} = \sum_{k=1}^{m} |x_{ik} - x_{jk}|
   $$
   Suma las distancias absolutas entre las coordenadas.

3. **Distancia Coseno**:
   $$
   d_{ij} = 1 - \frac{\sum_{k=1}^m x_{ik} x_{jk}}{\|x_i\| \|x_j\|}
   $$
   Mide la similitud angular entre dos puntos.

4. **Distancia de Mahalanobis**:
   $$
   d_{ij} = \sqrt{(x_i - x_j)^T S^{-1} (x_i - x_j)}
   $$
   Donde $ S $ es la matriz de covarianza. Es útil cuando los datos tienen correlaciones.

---

### 2. **Matriz de Distancia Inicial**

En el inicio del clustering jerárquico, cada punto de datos se considera como un clúster individual. Por lo tanto, la matriz $ D $ contiene las distancias entre cada par de puntos. Las entradas diagonales de la matriz $ D $ son cero, ya que la distancia de un punto consigo mismo es siempre 0:
$$
D =
\begin{bmatrix}
0 & d_{12} & d_{13} & \cdots & d_{1n} \\
d_{21} & 0 & d_{23} & \cdots & d_{2n} \\
d_{31} & d_{32} & 0 & \cdots & d_{3n} \\
\vdots & \vdots & \vdots & \ddots & \vdots \\
d_{n1} & d_{n2} & d_{n3} & \cdots & 0
\end{bmatrix}
$$

---

### 3. **Métodos de Enlace (Linkage)**

El agrupamiento jerárquico combina iterativamente los clústeres con menor disimilitud. El método de enlace determina cómo se calcula la distancia entre clústeres $ A $ y $ B $. Si $ A $ y $ B $ contienen múltiples puntos, las distancias pueden definirse de varias maneras:

1. **Enlace Simple (Single Linkage)**  
   La distancia entre dos clústeres se calcula como la distancia mínima entre un punto de $ A $ y un punto de $ B $:
   $$
   d(A, B) = \min_{x \in A, y \in B} d(x, y)
   $$
   **Ventaja**: Captura estructuras en forma de cadena.  
   **Desventaja**: Sensible al ruido y a los valores atípicos.

2. **Enlace Completo (Complete Linkage)**  
   La distancia entre clústeres es la máxima distancia entre un punto de $ A $ y un punto de $ B $:
   $$
   d(A, B) = \max_{x \in A, y \in B} d(x, y)
   $$
   **Ventaja**: Tiende a producir clústeres compactos.  
   **Desventaja**: Puede fragmentar clústeres grandes.

3. **Enlace Promedio (Average Linkage)**  
   La distancia se calcula como el promedio de todas las distancias entre puntos de $ A $ y $ B $:
   $$
   d(A, B) = \frac{1}{|A| \cdot |B|} \sum_{x \in A} \sum_{y \in B} d(x, y)
   $$
   **Ventaja**: Balance entre enlace simple y completo.  
   **Desventaja**: Puede ser lento para grandes clústeres.

4. **Enlace del Centroide (Centroid Linkage)**  
   La distancia se mide entre los centroides de los clústeres $ A $ y $ B $:
   $$
   d(A, B) = \| \mu_A - \mu_B \|
   $$
   Donde $ \mu_A $ y $ \mu_B $ son los centroides (medias) de los clústeres:
   $$
   \mu_A = \frac{1}{|A|} \sum_{x \in A} x
   $$

5. **Ward’s Method**  
   Minimiza la varianza total dentro de los clústeres. La distancia entre dos clústeres se calcula como el aumento en la suma de las varianzas al combinarlos:
   $$
   d(A, B) = \frac{|A| \cdot |B|}{|A| + |B|} \|\mu_A - \mu_B\|^2
   $$
   Este método favorece clústeres compactos y de tamaño similar.

---

### 4. **Actualización Iterativa de la Matriz de Distancias**

En cada paso del clustering jerárquico, los dos clústeres más cercanos (según la métrica de enlace) se combinan en un nuevo clúster. La matriz de distancias se actualiza eliminando las filas y columnas correspondientes a los clústeres originales y añadiendo una nueva fila/columna para el nuevo clúster.

Por ejemplo, si $ A $ y $ B $ se combinan en un nuevo clúster $ C $, las distancias entre $ C $ y otro clúster $ D $ se calculan según el método de enlace elegido.

---

### 5. **Construcción del Dendrograma**

El dendrograma es una representación visual de cómo se combinan los clústeres. Matemáticamente, se construye registrando la distancia $ d $ en la que se fusionan los clústeres $ A $ y $ B $ en cada paso. Estas distancias definen las alturas de las bifurcaciones en el dendrograma.

---

### Complejidad Computacional

1. **Cálculo inicial de la matriz de distancias**:  
   Complejidad $ O(n^2) $ para $ n $ puntos.  

2. **Actualización iterativa**:  
   Hay $ n-1 $ pasos, y en cada uno se actualizan las distancias entre clústeres, lo que puede implicar $ O(n^2) $ operaciones.  
   
   **Complejidad total**: $ O(n^3) $ en el peor caso, aunque con optimizaciones se puede reducir a $ O(n^2) $.  



### Ventajas del Agrupamiento Jerárquico

1. **No requiere predefinir $ K $**:  
   A diferencia de K-means, permite explorar la estructura completa de los datos y decidir posteriormente el número de clústeres.  

2. **Visualización intuitiva**:  
   Los dendrogramas facilitan la interpretación de las relaciones entre los datos.  

3. **Flexible con formas de clústeres**:  
   No asume que los clústeres tienen formas esféricas, lo que lo hace útil para datos más complejos.  

4. **Menos dependiente de la inicialización**:  
   Los resultados no dependen de una selección inicial aleatoria, como ocurre con K-means.  

---

### Limitaciones del Agrupamiento Jerárquico

1. **Escalabilidad**:  
   La complejidad computacional es $ O(n^2) $, lo que lo hace menos adecuado para grandes conjuntos de datos.  

2. **Sensibilidad al ruido**:  
   Puntos de datos atípicos o ruido pueden afectar significativamente la construcción del dendrograma.  

3. **Elección del método de enlace**:  
   Diferentes métodos de enlace pueden producir resultados distintos, lo que puede complicar la interpretación.  

---






```python
# Importamos las bibliotecas necesarias para el agrupamiento jerárquico
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.preprocessing import StandardScaler

# Preparamos los datos
X = data[['monto_gastado', 'frecuencia_compra']]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Creamos el modelo de agrupamiento jerárquico
linkage_matrix = linkage(X_scaled, method='ward')

# Visualizamos el dendrograma
plt.figure(figsize=(10, 7))
dendrogram(linkage_matrix)
plt.title('Dendrograma del Agrupamiento Jerárquico')
plt.xlabel('Índice de la muestra')
plt.ylabel('Distancia')
plt.show()

# Podemos cortar el dendrograma para obtener un número específico de clusters
from sklearn.cluster import AgglomerativeClustering

n_clusters = 4  # Usando el mismo número que en K-means para comparar
clustering = AgglomerativeClustering(n_clusters=n_clusters)
clusters = clustering.fit_predict(X_scaled)

# Visualizamos los resultados
plt.figure(figsize=(10, 6))
scatter = plt.scatter(data['monto_gastado'], data['frecuencia_compra'], 
                     c=clusters, cmap='viridis')
plt.xlabel('Monto Gastado ($)')
plt.ylabel('Frecuencia de Compra')
plt.title('Segmentación de Clientes usando Agrupamiento Jerárquico')
plt.colorbar(scatter, label='Cluster')
plt.show()

# Añadimos las etiquetas al DataFrame
data['Cluster_Jerarquico'] = clusters

# Analizamos los clusters resultantes
for i in range(n_clusters):
    cluster_data = data[data['Cluster_Jerarquico'] == i]
    print(f"\nCluster Jerárquico {i}:")
    print(f"Monto promedio: ${cluster_data['monto_gastado'].mean():.2f}")
    print(f"Frecuencia promedio: {cluster_data['frecuencia_compra'].mean():.2f}")
    print(f"Número de clientes: {len(cluster_data)}")
```

# Reducción de dimensionalidad

## Introducción a la Reducción de Dimensionalidad  

La reducción de dimensionalidad es una técnica fundamental en el aprendizaje automático y la ciencia de datos que busca transformar datos de alta dimensionalidad en representaciones de menor dimensión, manteniendo la mayor cantidad posible de información relevante. Este proceso no solo mejora la eficiencia computacional al trabajar con datos complejos, sino que también facilita su visualización e interpretación.  

En problemas de alta dimensionalidad, como el análisis de imágenes, textos o datos genómicos, los datos suelen ser redundantes o contener características irrelevantes. Esto puede llevar a problemas como el **"curse of dimensionality"**, donde el rendimiento de los modelos de aprendizaje automático se ve afectado negativamente. La reducción de dimensionalidad aborda estos desafíos al identificar patrones y relaciones subyacentes en los datos.  


### Propósitos clave:
1. **Eliminación de redundancias:** Identificar y remover características que están altamente correlacionadas.  
2. **Mejora del rendimiento:** Reducir la complejidad del modelo y evitar el sobreajuste.  
3. **Facilitación de la visualización:** Representar datos complejos en 2D o 3D para análisis exploratorio.  
4. **Captura de información semántica:**  Poder extraer representaciones e información semantica de datos de alta complejidad o dimensionalidad.





### Aplicaciones comunes:
- **Visualización de datos:** Comprender relaciones en datasets complejos como imágenes, texto o genomas.  
- **Preprocesamiento:** Reducir la dimensionalidad antes de aplicar algoritmos como clustering o clasificación.  
- **Compresión de datos:** Almacenar información relevante en menos dimensiones sin una pérdida significativa.  
- **Comprensión de los datos:** Trabajar a nivel semántico permite hacer comparaciones que capturen esa información


### Métodos destacados:
1. **Análisis de Componentes Principales (PCA):**  
   Una técnica lineal que proyecta los datos en un nuevo espacio utilizando componentes ortogonales que explican la máxima varianza. Es ideal para datos donde las relaciones son aproximadamente lineales.  

2. **t-Distributed Stochastic Neighbor Embedding (t-SNE):**  
   Un método no lineal diseñado para la visualización de datos de alta dimensionalidad. Optimiza la representación en 2D o 3D preservando las relaciones de vecindad locales.  

3. **Uniform Manifold Approximation and Projection (UMAP):**  
   Similar a t-SNE, pero con un enfoque más robusto y eficiente en la reducción de dimensionalidad. Mantiene tanto las estructuras locales como globales de los datos.  


### Análisis de Componentes Principales (PCA)

El **Análisis de Componentes Principales (PCA, por sus siglas en inglés)** es una técnica estadística ampliamente utilizada en el aprendizaje automático y la ciencia de datos para reducir la dimensionalidad de los datos mientras se preserva la mayor cantidad de información posible. PCA transforma los datos originales en un nuevo sistema de coordenadas donde las dimensiones (llamadas componentes principales) están ordenadas según la varianza explicada.  



### **Fundamento Matemático del PCA**  

1. **Estandarización de los datos:**  
   Para evitar que las características con mayor escala dominen el análisis, los datos son estandarizados:  
   $
   z_{ij} = \frac{x_{ij} - \mu_j}{\sigma_j}
   $  
   Donde $ x_{ij} $ es el valor original, $ \mu_j $ es la media de la característica $ j $, y $ \sigma_j $ es su desviación estándar. Esto asegura que todas las características tengan una media de 0 y una varianza de 1.  

2. **Cálculo de la matriz de covarianza:**  
   La matriz de covarianza captura las relaciones lineales entre las características. Para un conjunto de datos $ X $ con $ m $ características, la matriz de covarianza $ \Sigma $ es:  
   $
   \Sigma = \frac{1}{n-1} X^T X
   $  
   Donde $ n $ es el número de muestras.  

3. **Cálculo de los autovalores y autovectores:**  
   Los autovalores ($ \lambda $) y autovectores ($ v $) de la matriz de covarianza se calculan resolviendo la ecuación característica:  
   $
   \Sigma v = \lambda v
   $  
   - Los **autovalores** representan la varianza explicada por cada componente.  
   - Los **autovectores** son las direcciones (componentes principales) en el espacio de las características que maximizan la varianza.  

4. **Proyección de los datos:**  
   Una vez identificados los $ k $ componentes principales, los datos originales se proyectan en este nuevo subespacio:  
   $
   Z = X W_k
   $  
   Donde $ W_k $ es la matriz formada por los $ k $ autovectores principales seleccionados.  



### **Interpretación de los Componentes Principales**  

- **Primera Componente Principal (PC1):** Captura la mayor varianza posible en los datos. Es la dirección en la que los puntos están más dispersos.  
- **Segunda Componente Principal (PC2):** Ortogonal a PC1 y captura la siguiente mayor varianza posible, y así sucesivamente.  

Cada componente es una combinación lineal de las características originales, lo que permite identificar las direcciones más importantes del espacio de datos.  



### **Ventajas del PCA**  
1. **Reducción de ruido:** Al enfocarse en los componentes principales, PCA elimina dimensiones menos informativas.  
2. **Menor complejidad computacional:** Trabajar con menos dimensiones mejora la eficiencia de los modelos de aprendizaje automático.  
3. **Visualización de datos:** PCA facilita la representación gráfica de datos en 2D o 3D, útil para análisis exploratorio.  



### **Limitaciones del PCA**  
1. **Linealidad:** PCA asume que las relaciones entre las variables son lineales, lo que lo limita en casos con relaciones no lineales.  
2. **Pérdida de interpretabilidad:** Los componentes principales son combinaciones de características originales, lo que puede dificultar su interpretación.  
3. **Sensibilidad a la escala:** Aunque se recomienda estandarizar los datos, PCA sigue siendo sensible a la escala de las variables.  



### **Aplicaciones del PCA**  
1. **Procesamiento de imágenes:** Reducción de dimensionalidad en datos de píxeles para clasificar imágenes o realizar reconocimiento facial.  
2. **Genómica:** Análisis de datos genéticos para identificar patrones significativos en bases de datos de alta dimensionalidad.  
3. **Análisis de datos financieros:** Identificación de factores subyacentes en series temporales de datos financieros.  
4. **Preprocesamiento de datos:** Reducción de dimensionalidad antes de aplicar clustering o aprendizaje supervisado.  



### **Varianza Explicada y Selección de Componentes**  

La cantidad de componentes principales seleccionados depende de la proporción de varianza explicada que se desee preservar. Esta proporción puede calcularse acumulativamente como:  
$
\text{Varianza acumulada} = \frac{\sum_{i=1}^k \lambda_i}{\sum_{i=1}^m \lambda_i}
$  
Donde $ k $ es el número de componentes seleccionados y $ m $ es el número total de características.  

Generalmente, se eligen los componentes que explican entre el 90% y el 95% de la varianza total.

### **t-Distributed Stochastic Neighbor Embedding (t-SNE): Una Exploración Profunda**  

El **t-Distributed Stochastic Neighbor Embedding (t-SNE)** es una técnica de reducción de dimensionalidad no lineal diseñada específicamente para visualizar datos de alta dimensionalidad en un espacio de baja dimensionalidad (generalmente 2D o 3D). A diferencia de métodos lineales como PCA, t-SNE se centra en preservar las relaciones de proximidad local entre puntos, optimizando una representación que mantenga la estructura de los datos en términos de vecindad. Es una herramienta poderosa para entender patrones en datos complejos.  



### **Formalización del t-SNE**  

El t-SNE basa su funcionamiento en la comparación de distribuciones de probabilidades que representan relaciones de vecindad en el espacio de alta y baja dimensionalidad. Su objetivo es minimizar la discrepancia entre estas distribuciones utilizando una métrica de divergencia de probabilidad.

1. **Medición de proximidad en el espacio de alta dimensionalidad:**  
   Para cada par de puntos $ x_i $ y $ x_j $ en el espacio original, t-SNE calcula una probabilidad condicional $ p_{j|i} $, que mide la similitud entre los puntos basada en una distribución gaussiana centrada en $ x_i $:  
   $
   p_{j|i} = \frac{\exp{\left(-\frac{\| x_i - x_j \|^2}{2\sigma_i^2}\right)}}{\sum_{k \neq i} \exp{\left(-\frac{\| x_i - x_k \|^2}{2\sigma_i^2}\right)}}
   $  
   Donde $ \sigma_i $ es un parámetro ajustable que controla la "percepción" de vecindad de $ x_i $.  

   La probabilidad simétrica $ p_{ij} $ se define como:  
   $
   p_{ij} = \frac{p_{j|i} + p_{i|j}}{2n}
   $  
   Aquí, $ n $ es el número total de puntos en el conjunto de datos.  

2. **Medición de proximidad en el espacio de baja dimensionalidad:**  
   En el espacio reducido (2D o 3D), t-SNE modela las relaciones entre puntos $ y_i $ y $ y_j $ utilizando una distribución t de Student con un grado de libertad (lo que produce colas más largas que una gaussiana):  
   $
   q_{ij} = \frac{\left(1 + \| y_i - y_j \|^2\right)^{-1}}{\sum_{k \neq l} \left(1 + \| y_k - y_l \|^2\right)^{-1}}
   $  

3. **Minimización de la divergencia entre distribuciones:**  
   El objetivo de t-SNE es minimizar la discrepancia entre las distribuciones $ p_{ij} $ (alta dimensionalidad) y $ q_{ij} $ (baja dimensionalidad) usando la **divergencia de Kullback-Leibler (KL)**:  
   $
   C = \sum_{i \neq j} p_{ij} \log{\frac{p_{ij}}{q_{ij}}}
   $  

   La minimización de esta función de costo se realiza mediante un algoritmo de optimización como Gradiente Descendente.  


### **Etapas Clave del Algoritmo t-SNE**  

1. **Calibración del parámetro de vecindad ($ \sigma_i $):**  
   t-SNE utiliza un parámetro llamado **perplejidad**, que controla la cantidad efectiva de vecinos considerados. Este parámetro se ajusta automáticamente para determinar el valor óptimo de $ \sigma_i $.  

2. **Inicialización de puntos en el espacio reducido:**  
   Los puntos en el espacio de baja dimensionalidad se inicializan aleatoriamente o usando PCA como paso preliminar para una convergencia más rápida.  

3. **Optimización iterativa:**  
   El algoritmo ajusta las posiciones de los puntos en el espacio reducido para minimizar la función de costo $ C $.  



### **Ventajas del t-SNE**  

1. **Preservación de la estructura local:**  
   t-SNE captura eficazmente las relaciones de vecindad entre puntos cercanos, lo que lo hace ideal para identificar clústeres y patrones locales.  

2. **Visualización efectiva:**  
   Es particularmente útil para representar gráficamente conjuntos de datos de alta dimensionalidad de forma comprensible en 2D o 3D.  

3. **Adaptabilidad:**  
   El parámetro de perplejidad permite ajustar el alcance de la vecindad considerada, adaptándose a diferentes tamaños de conjuntos de datos.  



### **Limitaciones del t-SNE**  

1. **Costoso computacionalmente:**  
   t-SNE tiene una complejidad de $ O(n^2) $, lo que lo hace ineficiente para conjuntos de datos muy grandes.  

2. **Escalabilidad limitada:**  
   Para datos masivos, técnicas como t-SNE acelerado (Barnes-Hut t-SNE) son necesarias para mejorar la velocidad.  

3. **Pérdida de estructura global:**  
   Aunque preserva relaciones locales, t-SNE puede distorsionar las relaciones globales entre clústeres.  

4. **Sensibilidad a parámetros:**  
   El rendimiento depende críticamente de parámetros como la perplejidad y la inicialización.

# Embeddings

```python
# Importamos las librerías necesarias
import torch
import torch.nn as nn
from collections import Counter
import PyPDF2
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Extraemos texto del PDF
def extraer_texto_pdf(ruta_pdf):
    with open(ruta_pdf, 'rb') as archivo:
        lector = PyPDF2.PdfReader(archivo)
        texto = ""
        for pagina in lector.pages:
            texto += pagina.extract_text()
    return texto

# Ruta del PDF (ajustar según necesidad)
ruta_pdf = "text/BrainLIkeInference.pdf"  
texto = extraer_texto_pdf(ruta_pdf)

# Preprocesamiento básico
texto = texto.lower()  # Convertimos a minúsculas
palabras = texto.split()  # Tokenización simple por espacios

# Creamos un vocabulario
vocab = Counter(palabras)
palabra_a_idx = {palabra: idx for idx, (palabra, _) in enumerate(vocab.most_common())}
idx_a_palabra = {idx: palabra for palabra, idx in palabra_a_idx.items()}

# Convertimos el texto a índices
indices = [palabra_a_idx[palabra] for palabra in palabras]

# Parámetros del embedding
vocab_size = len(vocab)
embedding_dim = 100  # Aumentamos dimensionalidad para mejor representación

# Creamos una capa de embedding
embedding = nn.Embedding(num_embeddings=vocab_size, embedding_dim=embedding_dim)

# Convertimos los índices a tensor
indices_tensor = torch.LongTensor(indices)

# Obtenemos los embeddings
word_embeddings = embedding(indices_tensor)

# Función para búsqueda semántica
def busqueda_semantica(query, n_resultados=5):
    # Procesamos la query
    query = query.lower().split()
    query_indices = [palabra_a_idx[p] for p in query if p in palabra_a_idx]
    if not query_indices:
        return []
    
    # Obtenemos embedding de la query
    query_tensor = torch.LongTensor(query_indices)
    query_embedding = embedding(query_tensor).mean(dim=0).detach().numpy()
    
    # Calculamos similitud con todas las palabras
    todas_embeddings = embedding.weight.detach().numpy()
    similitudes = cosine_similarity([query_embedding], todas_embeddings)[0]
    
    # Obtenemos los más similares
    mas_similares = np.argsort(similitudes)[-n_resultados:][::-1]
    return [(idx_a_palabra[idx], similitudes[idx]) for idx in mas_similares]

# Ejemplo de búsqueda
query = "inteligencia artificial"
resultados = busqueda_semantica(query)
print(f"\nBúsqueda semántica para: '{query}'")
for palabra, similitud in resultados:
    print(f"Palabra: {palabra}, Similitud: {similitud:.4f}")
```
